## 3. Производная

Производная – скорость изменения функции.  
Определение – предел приращения функции к приращению ее аргумента при стремлении приращения аргумента к 0:

$$f'(x) = lim_{\Delta x \rightarrow 0} \frac{\Delta f(x)}{\Delta(x)}$$

```python
x = np.linspace(-10, 10, 1000)
y = x ** 2
y_prime = 2 * x
fig, ax = plt.subplots()
ax.plot(x, y)
ax.plot(x, y_prime)
```

```python
from IPython.display import Image
Image('derivatives.png')
```

Нахождение значения функции в окрестности $x_0$ через производную:

$$f'(x_0) \sim \frac{y - y_0}{x - x_0} \implies y - y_0 \sim f'(x_0)(x - x_0) \implies y \sim y_0 + f'(x_0)(x - x_0)$$

## 4. Монотонность и выпуклость

### Монотонность и выпуклость

Монотонная функция – все время либо не убывает, либо не возрастает. Если всегда либо растет, либо убывает, то функция – строго монотонная.  
$e^x$ – строго монотонная.  
$x^2$ – не монотонная.

Выпуклая функция - та, для которой выполняется неравенство Йенсена:
$$f(t\cdot x_1 + (1 - t)\cdot x_2) \leq t \cdot f(x_1) + (1 - t) \cdot f(x_2),$$
$$x_1,\space x_2 \space – \space значения \space аргумента \space функции, t \in [0, 1]$$

Если человеческим языком, то к любой точке функции можно провести касательную и вся функция будет находиться только с одной стороны от касательной. Тогда функция выпуклая.

### Нули производной

Нули производной – это точки экстремума (максимума или минимума) функции, или точки ее перегиба.

```python
x = np.linspace(-10, 10, 1000)
y = x ** 2
fig, ax = plt.subplots()
ax.plot(x, y)
y_min = np.min(y)
x_min_pos = list(y).index(y_min)
x_min = np.array(x)[x_min_pos]
ax.plot(x_min, y_min, 'ro')
ax.text(x_min, y_min + 10, 'точка минимума для x^2');
```

```python
x = np.linspace(-10, 10, 1000)
y = x ** 3
fig, ax = plt.subplots()
ax.plot(x, y)
ax.plot(0, 0, 'ro')
ax.text(0, 0 - 200, 'точка перегиба для x^3');
```

### Периоды монотонности

Промежутки между нулями производной – периоды монотонности

```python
x = np.linspace(-10, 10, 1000)
y = 3 * x ** 3 - 2 * x ** 2 + x - 5
y_prime = 9 * x ** 2 - 4 * x + 1
fig, ax = plt.subplots()

ax.plot(x, y, label='f')
ax.plot(x, y_prime, label='f_prime')
zero_prime_coord = np.where(np.isclose(y_prime, 0, atol=1e-00))[0]
for x_coord in zero_prime_coord:
    ax.plot(x[x_coord], y_prime[x_coord], 'ro')

plt.legend();
```

```python
np.where(np.isclose(y_prime, 0, atol=1e-00))
```

```python
x = np.linspace(-10, 10, 1000)
y = np.sin(x)
y_prime = np.cos(x)
fig, ax = plt.subplots(figsize=(10, 8))
ax.plot(x, y, label='f')
ax.plot(x, y_prime, label='f prime')
zero_prime_coord = np.where(np.isclose(y_prime, 0, atol=1e-02))[0]
for x_coord in zero_prime_coord:
    ax.plot(x[x_coord], y_prime[x_coord], 'ro')
plt.legend();
```

### Частные производные

Для функции от нескольких переменных можно считать производную двумя способами – частные производные и градиент.  
При частной производной мы считаем все переменные кроме одной – константами, по оставшейся дифференцируем.

$$f(x, y) = x^2 \cdot e^y + 2y^3 \cdot x$$
$$f'_x = 2x \cdot e^y + 2y^3$$
$$f'_y = x^2 \cdot e^y + 6y^2 \cdot x$$

Второй способ – градиент. Берем все частные производные, составляем вектор значений из них:

$$\nabla f(x, y) = (2x \cdot e^y + 2y^3, x^2 \cdot e^y + 6y^2 \cdot x)$$

Функция растёт быстрее всего в направлении градиента, убывает в противоположном направлении.  
Это значит, что для минимизации функции можно делать [маленькие шаги](https://en.wikipedia.org/wiki/Learning_rate) в направлении антиградиента, см. [стохастический градиентный спуск](http://www.machinelearning.ru/wiki/index.php?title=%D0%A1%D1%82%D0%BE%D1%85%D0%B0%D1%81%D1%82%D0%B8%D1%87%D0%B5%D1%81%D0%BA%D0%B8%D0%B9_%D0%B3%D1%80%D0%B0%D0%B4%D0%B8%D0%B5%D0%BD%D1%82%D0%BD%D1%8B%D0%B9_%D1%81%D0%BF%D1%83%D1%81%D0%BA).

Гессиан - матрица вторых производных (по всем парам, поэтому матрица, а не вектор).  
Если она положительно определена, а градиент равен нулю, то это точка минимума. Используется в оптимизации, см. [метод Ньютона](https://ru.wikipedia.org/wiki/%D0%9C%D0%B5%D1%82%D0%BE%D0%B4_%D0%9D%D1%8C%D1%8E%D1%82%D0%BE%D0%BD%D0%B0).
