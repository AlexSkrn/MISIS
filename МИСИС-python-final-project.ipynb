```python
import soundfile as sf  #  библиотеку для работы с wav
import numpy as np
import scipy.fftpack    # библиотека для выполнения преобразования Фурье
```

```python
# Сигналу (в нашем случае звуку) можно сопоставить его спектр — это коэффициенты преобразования 
# Фурье на небольших промежутках (мы будем брать промежутки по 50 мс). 

# Для того, чтобы звук воспринимался ниже или выше, нужно «сместить» соответствующим образом 
# коэффициенты разложения Фурье. Например, если мы хотим повысить звук в два раза 
# (то есть перейти к следующей октаве), мы должны «переложить» значения из коэффициента 1 
# в коэффициент 2, из 2 в 4, из 3 в 6 и так далее. После того, как частоты поменяны, нам нужно 
# выполнить обратное преобразование Фурье и получить снова тот же сигнал, но уже повышенный на октаву.

# Из теории музыки известно, что переход на октаву выше соответствует увеличению частоты 
# звука ровно в два раза. Соответственно, для перехода на октаву ниже частоты нужно уменьшить в два раза.
# Следующая функция делает ровно это: понижает или повышает частоты в зависимости от 
# множителя mult (например, при mult=2 произойдёт понижение на одну октаву, 
# а при mult=1/2 высота будет повышена на октаву):
def transform(x, mult):
    n = len(x)
    y = np.zeros([n], dtype=np.cdouble)    # y - преобразованный сигнал
    for i in range(0, n // 2):
        k = int(i / mult)            # k - откуда переложить, i - куда переложить
        if k >= n / 2:
            continue
        y[i] = x[k]                                     # переставляем частоты
        y[n - i - 1] = x[n - k - 1]
    return y
```

```python
# следующая функция проходит по всем промежуткам по 50 мс и преобразовывает сигнал. 
# Сначала она запускает преобразование Фурье на каждом из промежутков, затем переставляет частоты, 
# используя предыдущую функцию, и выполняет обратное преобразование Фурье.
def process_frames(name, window):
    sig, samplerate = sf.read(name)                     # читаем файл в память в виде numpy-массива
    if len(sig.shape) == 2:
        sig = sig[:, 0]
    step = int(samplerate * window)                     # переводим окно из миллисекунд в сигналы
    data, embeddings = [], []
    for i in range(0, sig.shape[0] - step, step):
        y = sig[i : i + step]
        y = scipy.fftpack.fft(y)  # прямое преобразование Фурье
        # функция embeddings.append понадобится для визуализации данных в четвертом проектном задании
        embeddings.append(np.abs(y))  
        y = transform(y, 1/2**(4/12))  # переставляем частоты
        y = scipy.fftpack.ifft(y)      # производим обратное преобразование
        data += list(map(lambda x: x.real, y))
    return data, samplerate
    # sf.write('test.wav', data, samplerate)  # записываем результат в файл
```
