````python
def encrypt(string):
   alf = ("а","б","в","г","д","е","ё","ж","з","и","й","к","л","м","н","о","п","р","с","т","у","ф","х","ц","ч","ш","щ","ъ","ы","ь","э","ю","я")
   new = []
   for i in string:
       if i not in alf:
           new.append(i)
       else:
           if  i == "я":
               new.append("а")
           else:
               new.append(alf[alf.index(i)+1])
   return "".join(new)
word = input('Введите слово для шифрования')
print(encrypt(word))
````

Алгоритм пузырьковой сортировки:

1. Последовательно попарно сравниваем элементы, если порядок в паре неверный, то выполняется обмен местами между элементами.
2. После того, как первый круг пройден, действие повторяется снова до тех пор, пока все элементы не будут стоять в верном положении.

```

            if элемент A[j] > A[j+1] then:
                temp = A[j]

                A[j+1] = temp
```

Алгоритм сортировки слиянием:

1. Данные разбиваются на примерно равные части.
2. Сортируется каждая из частей в отдельности.
3. Две отсортированные части сливаются вместе, и далее элементы выстраиваются в нужной последовательности.

https://www.youtube.com/watch?v=OGzPmgsI-pQ&list=PLqM7alHXFySHrGIxeBOo4-mKO4H8j2knW&index=3

Сортировка вставками — сортировка, в которой элементы просматриваются по одному и ставятся на место в соответствии с уже упорядоченным массивом.

Алгоритм сортировки вставками:

1. Сравниваем первые два элемента, сортируем их.
2. Выбираем третий элемент и находим его позицию между первыми двумя.
3. Далее выполняем аналогичные действия со всеми последующими элементами.
4. Таким образом, мы постепенно добавляем больше элементов в уже отсортированный список, помещая их в правильное положение.

````python
# Y = input().split()
Y = [19,2,31,35,30,11,121,27]  # list(map(int, Y))
for i in range(1, len(Y)):
    print(Y)
    b = Y[i]
    j = i-1
    while (j >= 0) and (b < Y[j]):
        Y[j+1] = Y[j]
        j-=1
    Y[j+1] = b
    
print(Y)
````

Этот алгоритм сортировки является усовершенствованным вариантом сортировки вставками. Его идея заключается в том, что сравниваются элементы, стоящие не только рядом, но и на определённом расстоянии друг от друга.

Алгоритм сортировки Шелла:

1. Сравниваются два элемента последовательности, находящиеся друг от друга на некотором расстоянии $d = N/2$, где $N$ — общее число элементов, которые в случае необходимости меняются местами.
2. Расстояние уменьшается до $d/2$, и сравниваются все элементы, между которым расстояние равно этому значению. Если необходимо, делается обмен местами между элементами.
3. Такой алгоритм продолжается до тех пор, пока $d !=1$, при $d=1$ проход по массиву осуществляется последний раз.

### Сортировка выбором — это гибрид между пузырьковой и сортировкой вставками.

## двоичный поиск

- вещественный двоичный поиск;

ЕЩЕСТВЕННЫЙ ДВОИЧНЫЙ ПОИСК

Общая идея метода: найти такое пороговое значение, где «хорошие» числа становятся «плохими». Хорошее или плохое число определяется некоторым заданным критерием. Например, у нас есть веревка, которая при одном значении массы груза выдержит его, а при другом — порвется. И необходимо найти пороговое значение, при котором веревка выдержит, а при его увеличении — надо идти за новой верёвкой.

Для решения такой задачи будет использоваться вещественный двоичный поиск.

**Пример**: Семья решила заказать домой новый комод, и им нужно решить, сколько потребуется грузчиков. Им сказали, что один грузчик сможет донести груз массой 10 кг, а массой 12 кг уже поднять не сможет. И нужно определить, какова максимальная масса груза, который сможет поднять один человек.

Задача будет решена вещественным двоичным поиском, т.к. масса в кг может быть, например, 10.45, а это число – вещественное.

ЦЕЛОЧИСЛЕННЫЙ ДВОИЧНЫЙ ПОИСК

Сформулируем общую идею аналогично вещественному двоичному поиску: в этом алгоритме необходимо выполнить поиск такого числа, которое будет являться границей между «хорошими» и «плохими» числами. Это число должно быть целым. 

Причем, если l - “хорошее”, а r - “плохое”, то l - r = 1.

__Алгоритм бинарного поиска:__

Шаг 1. Найти индекс среднего элемента массива middle = (high-low)//2

When we're talking about hash tables, we can define a __"load factor"__:

__Load Factor = Number of Entries / Number of Buckets__

The purpose of a load factor is to give us a sense of how "full" a hash table is. For example, if we're trying to store 10 values in a hash table with 1000 buckets, the load factor would be 0.01, and the majority of buckets in the table will be empty. We end up wasting memory by having so many empty buckets, so we may want to rehash, or come up with a new hash function with less buckets. We can use our load factor as an indicator for when to rehash—as the load factor approaches 0, the more empty, or sparse, our hash table is.

On the flip side, the closer our load factor is to 1 (meaning the number of values equals the number of buckets), the better it would be for us to rehash and add more buckets. Any table with a load value greater than 1 is guaranteed to have collisions.


For the load factor, you should divide the number of values by the number of buckets. There are 100 values, as stated in the question, and 100 buckets (0 to 99). Thus, 100/100 = 1!

````python
# solution
class HashTable(object):
    def __init__(self):
        self.table = [None]*10000

    def store(self, string):
        hv = self.calculate_hash_value(string)
        if hv != -1:
            if self.table[hv] != None:
                self.table[hv].append(string)
            else:
                self.table[hv] = [string]

    def lookup(self, string):
        hv = self.calculate_hash_value(string)
        if hv != -1:
            if self.table[hv] != None:
                if string in self.table[hv]:
                    return hv
        return -1

    def calculate_hash_value(self, string):
        value = ord(string[0])*100 + ord(string[1])
        return value
````

````python
def msort(a):
  res = []
  n = len(a)
  if n < 2:
    return a
  mid = n//2
  L, R = msort(a[:mid]), msort(a[mid:])
  i = j = 0
  while i < len(L) and j < len(R):
    if L[i] > R[j]:
      res.append(R[j])
      j += 1
    else:
      res.append(L[i])
      i += 1
  res += L[i:]
  res += R[j:]
  return res  
````

````python
a_list = [54, 26, 93, 17, 77, 31, 44, 55, 20]
# a_list = [2, 1]

def isort(x):
  for idx, val in enumerate(x):
    pos = idx
    while pos > 0 and x[pos-1] > val:
      x[pos] = x[pos-1]
      pos -= 1
    x[pos] = val
  return None

isort(a_list)
print(a_list)
````

````python
a_list = [54, 26, 93, 17, 77, 31, 44, 55, 20]
def insertion_sort(a_list):
  for index in range(1, len(a_list)):
    current_value = a_list[index] 
    position = index
    while position > 0 and a_list[position - 1] > current_value:
      a_list[position] = a_list[position - 1]
      position = position - 1
    a_list[position] = current_value
a_list = [54, 26, 93, 17, 77, 31, 44, 55, 20]
insertion_sort(a_list)
print(a_list)
````
